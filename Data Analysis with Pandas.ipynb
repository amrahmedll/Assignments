```python
#1 average is the mean?
avg_value_method1 = data['Tenure_Months'].mean()
print("the average is(method1):", avg_value_method1)

#OR

sumval = data['Tenure_Months'].sum()
countval = data['Tenure_Months'].count()
avg_value_method2 = sumval / countval
print("the average is(method2):", avg_value_method2)
```
